```python
import google.generativeai as genai
import os

# Google Generative AI API 키를 설정합니다.
genai.configure(api_key="발급받은_API_키")

# "gemini-1.5-flash" 모델을 사용해 응답을 생성합니다.
model = genai.GenerativeModel("gemini-1.5-flash")

# "서울 명소는?" 질문을 모델에 전달하여 응답을 생성합니다.
response = model.generate_content("서울 명소는?")

# 모델의 응답을 출력합니다.
print(response.text)
```
